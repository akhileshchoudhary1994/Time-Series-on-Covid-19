# Exponential Smoothing 

```python
exp_smoothing=[Kerala1.Confirmed[0]]
alpha=0.5
for i in range(len(Kerala1.Confirmed)-1):
    e=alpha*Kerala1.Confirmed[i+1]+(1-alpha)*Kerala1.Confirmed[i]
    exp_smoothing.append(e)
```

```python
plt.figure()
plt.plot(range(142),Kerala1['Confirmed'])
plt.plot(exp_smoothing)
plt.show()
```

# Method of Curve Fitting

```python
import numpy as np
import statsmodels.formula.api as smf
```

```python
var='Confirmed'

regression=pd.DataFrame(Kerala1[var][1:])
regression['t']=range(1,regression.shape[0]+1)
```

```python
regression
```

```python
regression[var].plot()
```

Data Visiualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
# Seaborn scatter plot with regression line
sns.lmplot(x='t', y='Confirmed', data=regression, aspect=1.5, scatter_kws={'alpha':0.2})
```

```python
def get_formula(x, y, degree):
    
    formula = f'{y} ~ {x}'
    if degree == 1:
        return formula
    
    i = 2
    while i <= degree:
        formula += f' + I({x}**{i})'
        i += 1 
        
    return formula
```

```python
get_formula('t', 'Confirmed', 6)
```

```python
def ols_fit(x, y, data, degree):
    formula = get_formula(x, y, degree)
    res = smf.ols(formula, data=data).fit()
    
    return res
```

```python
def check_sig(ols_res, x, degree, sig):
    if degree == 1:
        pval = res.pvalues[x]
    else:
        pval = res.pvalues[f'I({x} ** {degree})']
        
    if pval < sig:
        return True
    
    return False
    
```

```python
res = ols_fit('t', 'Confirmed', regression, 2)
print(res.summary())
```

```python
models = {}
sigs = {}
```

```python
degree = 1
while True:
    res = ols_fit('t', 'Confirmed', regression, degree)
    models[degree] = res
    if not check_sig(res, 't', degree, sig=0.05):
        sigs[degree] = False
        break
        
    sigs[degree] = True
    degree += 1
    
print(f'The best fitted polynomial trend is of degree {degree - 1}')
```

```python
var = 'Confirmed'
actual = regression[var]
def rmse(obs, mod):
    return ((obs - mod)**2).mean()**(1/2)
```

```python
fig, axn = plt.subplots(3, 1, figsize=(12, 12))
colors = ['red', 'green', 'blue']

for col, ax, (degree, mod) in zip(colors, axn, models.items()):
    ax.plot(regression.index, actual, alpha=0.5, color='black')
    pred = models[degree].predict(regression['t'])
    ax.plot(regression.index, pred, color=col)
    
    ax.text(0.1, 0.8,'RMSE = {:.2f}'.format(rmse(actual, pred)),
            horizontalalignment='center',
            verticalalignment='center',
            transform = ax.transAxes,
            fontsize=10, fontweight='bold')
    
    title = f'Degree {degree}'
    if sigs[degree]:
        title += '*'
    ax.set_title(title, fontweight='bold')

plt.tight_layout()
fig.suptitle(f'India Daily {var} COVID cases: Polynomial trend', 
             fontsize=20, fontweight='bold')

fig.subplots_adjust(top=0.9)
```
